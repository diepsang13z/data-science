# Nutrition Data Visualisation on 80 Cereal products
See: https://www.kaggle.com/datasets/crawford/80-cereals

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
sns.set_theme()
```

```python
root_cereal_df = pd.read_csv('./cereal.csv')
root_cereal_df.head()
```

```python
# Irrelevant fields
irrelevant_fields = ['shelf', 'weight', 'cups', 'rating']
analysis_cereal_df = root_cereal_df.drop(irrelevant_fields, axis=1)
analysis_cereal_df.head()
```

```python
# Get correlation data
cereal_corr = analysis_cereal_df.drop(['name', 'mfr', 'type'], axis=1).corr()
cereal_corr
```

```python
# Build a matrix of booleans with the same shape as our data
ones_corr = np.ones_like(cereal_corr, dtype='bool')
ones_corr
```

```python
ones_corr.shape, cereal_corr.shape
```

```python
# Get only upper triangle matrix to draw heatmaps
mask = np.triu(ones_corr)
mask
```

```python
sns.heatmap(data=cereal_corr, mask=mask, cmap='Blues')
plt.show()
```

```python
# When removing the upper tri-angle, here are still two empty cells in our matrix (calories & vitamin)
adjusted_mask = mask[1:, :-1]
adjusted_cereal_corr = cereal_corr.iloc[1:, :-1]
```

```python
fig, ax = plt.subplots(figsize=(10, 8))

#That method uses HUSL colors, so you need hue, saturation, and lightness. 
#I used hsluv.org to select the colors of this chart.
custom_cmap = sns.diverging_palette(0, 230, 90, 60, as_cmap=True)

# Draw plot
heatmap_style = {
    'cmap': custom_cmap,
    'fmt': '.2f',
    'annot': True,
    'vmin': -1,
    'vmax': 1,
    'linecolor': 'white',
    'linewidths': 0.5,
}
sns.heatmap(data=adjusted_cereal_corr, mask=adjusted_mask, **heatmap_style)

# Uppercase labels
xticks = [i.upper() for i in adjusted_cereal_corr.columns]
yticks = [i.upper() for i in adjusted_cereal_corr.index]

ax.set_xticklabels(xticks, rotation=90, fontsize=13)
ax.set_yticklabels(yticks, rotation=0,  fontsize=13)

# Set plot title
title = 'CORRELATION MATRIX\nSAMPLED CEREALS COMPOSITION\n'
ax.set_title(title, loc='left', fontsize=18);

plt.show()
```
