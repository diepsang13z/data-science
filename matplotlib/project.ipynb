# Project: Data Visualization - California Cities

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
cities = pd.read_csv('./california_cities.csv')
```

```python
cities.head()
```

```python
# Extract data
lat, lot = cities['latd'], cities['longd']
population, area = cities['population_total'], cities['area_total_km2']
```

```python
# Plot using Pyplot API
plt.style.use('seaborn-v0_8')
plt.figure(figsize=(8, 6))

plt.scatter(lat, lot, c=np.log10(population), cmap='viridis', s=area, linewidth=0, alpha=0.7)

plt.xlabel('Longlatitude')
plt.ylabel('Lattitude')

plt.title('California Cities: Population and Area Distribution')
plt.axis('equal')
plt.colorbar(label='log$_{10}$(population)')
plt.clim(3, 7)

# Create a legend for cities'sizes
area_range = [50, 100, 300, 500]
for area in area_range:
    plt.scatter([], [], s=area, c='k', alpha=0.4, label=str(area) + 'km$^2$')
plt.legend(scatterpoints=1, labelspacing=1, title='City Area')
             
plt.show()
```
