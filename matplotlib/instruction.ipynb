## Pyplot API vs Object-Oriented API
- Pyplot API -> Quickly
- OO API -> Advanced

### Pyplot API

```python
x = np.linspace(0, 10, 1000)

plt.title('Sin(x) and Cos(x)')

plt.plot(x, np.sin(x), color='blue', label='sin(x)', linestyle='dashed')
plt.plot(x, np.cos(x), color='red', label='cos(x)')
plt.legend()

plt.xlabel('x variable')
plt.ylabel('F(x)')

plt.xlim([0, 9])
plt.axis('equal')

plt.show()
```

### Object-Oriented (OO) API

```python
x = [1, 2, 3, 4]
y= [11, 22, 33, 44]

fig, ax = plt.subplots(figsize=(5, 5))

ax.plot(x, y)
ax.set(title='A simple plot'
       , xlabel='x-axis'
       , ylabel='y-axis')

plt.show()
```

# Most common types of Matplotlib plots
- Line
- Scatter
- Bar
- Hist
- Subplots

```python
x = np.linspace(0, 10, 100)
```

## 1. Line Chart

```python
fig, ax = plt.subplots()
ax.plot(x, x**3)
plt.show()
```

## 2. Scatter Chart

```python
plt.scatter(x, np.exp(x))
plt.show()
```

```python
rng = np.random.RandomState(0)

x = rng.randn(100)
y = rng.randn(100)

size = 1000 * rng.randn(100)
color = 1000 * rng.randn(100)

fig, ax = plt.subplots()
my_plot = ax.scatter(x, y, s=size, c=color, cmap='viridis', alpha=0.7)
fig.colorbar(my_plot)

plt.show()
```

## 3. Bar chart
- Vertical
- Horizontal

```python
soft_drink_prices = {
    'Coke': 10,
    'Pessi': 12,
    'Sprite': 8,
}
```

```python
fig, ax = plt.subplots()
ax.bar(soft_drink_prices.keys(), soft_drink_prices.values())
ax.set(title="BHX's Soft Drink Prices", ylabel='Price ($)')
plt.show()
```

```python
fig, ax = plt.subplots()
ax.barh(soft_drink_prices.keys(), soft_drink_prices.values())
ax.set(title="BHX's Soft Drink Prices", ylabel='Price ($)')
plt.show()
```

## 4. Histogram

```python
np.random.seed(42)
student_height = np.random.normal(170, 10, 250)
```

```python
plt.hist(student_height)
plt.show()
```

```python
fig, ax = plt.subplots()
ax.hist(student_height, bins=30)
plt.show()
```

## 5. Subplots
See: https://matplotlib.org/3.1.1/gallery/recipes/create_subplots.html

```python
fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(nrows=2
                      , ncols=2
                      , figsize=(10, 5))

ax1.plot(x, x/2, color='blue');
ax1.plot(x, x*2, color='red', linestyle='dashed');

np.random.seed(33)
ax2.scatter(np.random.random(10), np.random.random(10));

ax3.bar(soft_drink_prices.keys(), soft_drink_prices.values());
ax4.hist(student_height, bins=30);

plt.show()
```

```python
fig, ax = plt.subplots(nrows=2
                      , ncols=2
                      , figsize=(10, 5))

ax[0, 0].plot(x, x/2, color='blue');
ax[0, 0].plot(x, x*2, color='red', linestyle='dashed');

np.random.seed(33)
ax[0, 1].scatter(np.random.random(10), np.random.random(10));

ax[1, 0].bar(soft_drink_prices.keys(), soft_drink_prices.values());
ax[1, 1].hist(student_height, bins=30);

plt.show()
```
